```python
"""
Selecting the important columns
1)genres
2)id
3)keywords
4)title
5)overview
6)cast
7)crew
"""
movies = movies[['movie_id', 'title', 'overview', 'genres', 'cast', 'crew', 'keywords']]
movies.head(3)
```

```python
import ast
def convert(obj):
  L =  []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].head()
```

```python
movies['keywords'].head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#Creating a funtion for the cast column for selecting only top 3 actors or cast

def convertcast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L        
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies['cast']
```

```python
movies['crew'][0]
```

```python
#Fetching the main crew(director,etc) from the crew column
def fetch_director(obj):
  L =  []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
        L.append(i['name'])
        break
  return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew']
```

```python
movies['overview'][0]
```

```python
#Converting the string in overview column into list

movies['overview'] =  movies['overview'].apply(lambda x:x.split())
movies['overview']
```

```python
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ", "") for i in x])

movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
      y.append(ps.stem(i))

  return " ".join(y)   
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:11]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:11]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Spiderman')
```
